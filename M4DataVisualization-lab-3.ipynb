# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
df_hist = pd.read_sql_query(QUERY,conn)

df_hist.plot(kind='hist')

plt.title('histogram of ConvertedComp')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
        SELECT Age
        FROM master
        """
df_box = pd.read_sql_query(QUERY,conn)

df_box.plot(kind='box')

plt.title('box plot of Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
        SELECT Age,WorkWeekHrs 
        FROM master
        """
df_scatter = pd.read_sql_query(QUERY,conn)

df_scatter.plot(kind='scatter', x= 'Age', y='WorkWeekHrs')

plt.title('Age and WorkWeekHrs')

plt.show()
```

### Bubble Plots


WorkWeekHrsCreate a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import plotly.graph_objects as go
import plotly.express as px
import numpy as np

QUERY = """
        SELECT Age,WorkWeekHrs,CodeRevHrs
        FROM master
        """
df_bubble = pd.read_sql_query(QUERY,conn)
plt.scatter(df_bubble['WorkWeekHrs'],df_bubble['CodeRevHrs'],s=df_bubble['Age'],alpha=0.5)
plt.title('Bubble plot of workweek hours, codeRevHrs and age')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, count(DatabaseDesireNextYear)as count
FROM DatabaseDesireNextYear 
group by DatabaseDesireNextYear
order by count desc
"""
df_pie = pd.read_sql_query(QUERY,conn)

df_pie = df_pie.set_index('DatabaseDesireNextYear')

df_pie = df_pie.head(5)
df_pie['percent']= (df_pie['count']/df_pie['count'].sum())*100


print(df_pie)

df_pie['count'].plot(kind='pie', autopct='%1.1f%%') 

plt.show()

QUERY2 = """
SELECT LanguageDesireNextYear, count(LanguageDesireNextYear)as count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count desc
"""
df_lang = pd.read_sql_query(QUERY2,conn)

print("\nlanguages desired to learn next year:\n\n" , df_lang)

table_name = 'LanguageWorkedWith'  # the table you wish to describe


QUERY3 = """
SELECT LanguageWorkedWith, count(LanguageWorkedWith)as count
FROM LanguageWorkedWith
group by LanguageWorkedWith
order by count desc
"""
df_work = pd.read_sql_query(QUERY3,conn)

print("\nlanguages work with:\n\n" , df_work)

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df_stacked = pd.read_sql_query(QUERY,conn)

df_stacked = df_stacked.groupby('Age').median()

print(df_stacked)



```

```python
df_stacked.plot(kind='bar',stacked=True)
plt.title('Median workweek hrs and CodeRevHrs for the age group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_line = pd.read_sql_query(QUERY,conn)

df_line = df_line.groupby('Age').median()

print(df_line)

df_line.plot(kind='line')

plt.title('Median salary in 45-60 age group')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')

plt.show()

QUERY2 = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 25 AND 30
"""
df_line2 = pd.read_sql_query(QUERY2,conn)

df_line2 = df_line2.groupby('Age').median()

print(df_line2)

df_line2.plot(kind='line')

plt.title('Median salary in 25-30 age group')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
df_mb = pd.read_sql_query(QUERY,conn)
df_barh=df_mb['MainBranch'].value_counts().rename_axis('MainBranch').reset_index(name='Counts')

df_barh

plt.barh(df_barh['MainBranch'], df_barh['Counts'],align='center')
plt.title('Main branch')
plt.ylabel('Main branch')
plt.xlabel('count')
plt.show()

QUERY2 = """
SELECT DevType, count(DevType)as count
FROM DevType
group by DevType
order by count
"""
df_devtype = pd.read_sql_query(QUERY2,conn)
df_devtype

plt.barh(df_devtype['DevType'], df_devtype['count'],align='center')
plt.title('dev type')
plt.ylabel('dev type')
plt.xlabel('count')
plt.show()
```
